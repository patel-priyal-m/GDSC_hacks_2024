# MAPING OF TEMPERATURE DATA







```python
import pandas as pd
import folium
from folium.plugins import HeatMapWithTime

# Group the DataFrame by date and create a list of dataframes
grouped = df.groupby(df['DATE'])

df_list = [group for _, group in grouped]

# Create a list of heat_data, which will be used for the HeatMapWithTime
heat_data = [[[row['LAT'],row['LON'],row['TEMP']] for index, row in df.iterrows()] for df in df_list]

# Initialize the map
m = folium.Map([df['LAT'].mean(), df['LON'].mean()], zoom_start = 7)

# Add HeatMapWithTime to the map
HeatMapWithTime(heat_data, radius=5, gradient={0.4: 'blue', 0.7: 'lime', 0.9: 'orange', 1: 'red'}, min_opacity=0.2, max_opacity=0.5, use_local_extrema=True).add_to(m)

# Save the map to an HTML file
m.save(outfile='heatmap.html')

```

```python
# Group the DataFrame by date and create a list of dataframes
grouped = df.groupby(df['DATE'])

df_list = [group for _, group in grouped]

# Create a list of heat_data with date information
heat_data = [[[row['LAT'], row['LON'], row['TEMP']] for _, row in df.iterrows()] for df in df_list]

# Get the dates
dates = [str(df['DATE'].iloc[0]) for df in df_list]

# Initialize the map
m = folium.Map([df['LAT'].mean(), df['LON'].mean()], zoom_start=7)

# Add HeatMapWithTime to the map
HeatMapWithTime(heat_data, radius=5, gradient={0.4: 'blue', 0.7: 'lime', 0.9: 'orange', 1: 'red'}, min_opacity=0.2, max_opacity=0.5, use_local_extrema=True, index=dates).add_to(m)

# Save the map to an HTML file
m.save(outfile='heatmap.html')
```

```python
import plotly.express as px

# Convert 'DATE' column to string format
df['DATE'] = df['DATE'].astype(str)

# Create the animated heatmap
fig = px.density_mapbox(df, lat='LAT', lon='LON', z='TEMP', radius=10,
                        center=dict(lat=27, lon=-95), zoom=1,
                        mapbox_style="open-street-map", animation_frame='DATE')
                        # mapbox_style="white-bg", animation_frame='DATE')


fig.update_layout(
    mapbox_style="white-bg",
    mapbox_layers=[
        {
            "below": 'traces',
            "sourcetype": "raster",
            "sourceattribution": "United States Geological Survey",
            "source": [
                "https://basemap.nationalmap.gov/arcgis/rest/services/USGSImageryOnly/MapServer/tile/{z}/{y}/{x}"
            ]
        },
        {
            "sourcetype": "raster",
            "sourceattribution": "Government of Canada",
            "source": ["https://geo.weather.gc.ca/geomet/?"
                       "SERVICE=WMS&VERSION=1.3.0&REQUEST=GetMap&BBOX={bbox-epsg-3857}&CRS=EPSG:3857"
                       "&WIDTH=1000&HEIGHT=1000&LAYERS=RADAR_1KM_RDBR&TILED=true&FORMAT=image/png"],
        }
      ])

# Show the figure
fig.show()
```
